```python
import pandas as pd
import datetime
```

# Import and Clean Data

```python
data = pd.read_csv("dji.csv")
```

```python
data
```

```python
data = data.drop('epoch_time', axis= 1)
data
```

```python
data.info()
```

```python
data['date_time'] = data['date_time'].dt.date
data
```

```python
# data['date_time'] = pd.to_datetime(data['date_time'])
# data.info()
```

```python
data = data.rename(columns={'date_time': 'date'})
data.info()
```

```python
data
```

```python
data = data.set_index('date')
data
```

```python
data.info()
```

```python
data.index = pd.to_datetime(data.index)
```

```python
df = data.loc["1921-04-15": "2021-04-16", "close"].to_frame()
df
```

# Data Visualization

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
plt.style.use("seaborn")
```

```python
df.describe()
```

```python
df.plot(figsize = (25, 20), fontsize = 20)
plt.legend(fontsize = 15)
plt.show()
```

```python
df['daily_return'] = df.pct_change()
df
```

```python
df['daily_return'].isna().sum()
```

```python
df.dropna(inplace = True)
df
```

```python
df.plot(figsize = (18, 9), secondary_y = "daily_return", mark_right = True, fontsize = 15)
plt.show()
```

```python
df.close.plot(figsize = (25, 20), fontsize = 15)
df.daily_return.where(df.daily_return.ge(0)).plot(figsize = (18, 10), secondary_y = "daily_return", mark_right = True, fontsize = 15, color='green')
df.daily_return.where(df.daily_return.lt(0)).plot(figsize = (18, 10), secondary_y = "daily_return", mark_right = True, fontsize = 15, color='red')
plt.legend(fontsize = 20)
plt.show()
```
