## Backtesting a simple Momentum Strategy

__Positions__: <br>
+1: Investing in DJI (__long__ position) <br>
-1: Short Selling DJI (__short__ position)<br>
0: No position (__neutral__)
    

__Strategies__ <br>
-__Buy and Hold__ (Basic Strategy - passive): Initially Investing into DJI and do nothing (Position: +1 on any given day)<br>
-__Simple Momentum__ (active Strategy to be tested): <br>
__a) Investing (+1)__ into DJI tomorrow if today´s return was __positive__ <br>
__b) Short selling (-1)__ DJI tomorrow if today´s return was __negative__

```python
# The position in this case will be the position taken for the following day

df2["position"] = np.sign(df2["DJI_daily_return"])
df2
```

```python
# The strategy_return is the percent return that would have happened with our strategy/position. 
# It is shifted one row since it is the following day

df2["strategy_percent_return"] = df2["position"].shift() * df2["DJI_daily_return"]
df2
```

```python
# will now add a column that reflect the overall cumulitive percent from the strategy daily percent return if taken position

df2.strategy_percent_return.add(1, fill_value = 0).cumprod()
```

```python
# Normalize the cumulitive percent return to the first date to mirror starting value

df2["strategy_value_return"] = df2.strategy_percent_return.add(1, fill_value = 0).cumprod() * df2.iloc[0, 0]
df2
```

```python
df2[["DJI_close", "strategy_value_return"]].plot(figsize = (15, 10), fontsize = 15)
plt.title("Simple Momentum Strategy", fontsize = 20)
plt.legend(fontsize = 15)
plt.show()
```

```python
# 252 usual number of trading days in a year. Here is an exact calculation based on number of rows

print(2516/10)
```

```python
# anual return and risk average

def annualized_risk(returns):
    # create column mean and std
    summary = returns.agg(["mean", "std"]).T # .T is transpose columns and rows
    
    # rename mean column to return and multiply time trading days of year
    summary["Return"] = summary["mean"] * 251.6
    
    # rename std column as risk(volatility) and multiply times square root of trading days
    summary["Risk"] = summary["std"] * np.sqrt(251.6)
    
    #drop the mean and std columns
    summary.drop(columns = ["mean", "std"], inplace = True)
    return summary
    
```

```python
annualized_risk(df2[["DJI_daily_return", "strategy_percent_return"]])
```

## Backtesting a simple Contrarian Strategy

__Strategies__: <br>
-__Buy and Hold__ (Basic Strategy): Initially Investing into DJI and do nothing (Position: +1 on all days)

-__Simple Contrarian__ (Strategy to be tested): <br>
__a) Short Selling (-1)__ DJI tomorrow if today´s return was __positive__ <br>
__b) Investing (+1)__ into DJI tomorrow if today´s return was __negative__

```python
df3 = df2.copy()
df3
```

```python
df3[['position','strategy_percent_return']] = df3[['position','strategy_percent_return']].apply(lambda x: x * -1)
df3
```

```python
df3["strategy_value_return"] = df3.strategy_percent_return.add(1, fill_value = 0).cumprod() * df3.iloc[0, 0]
df3
```

```python
df3[["DJI_close", "strategy_value_return"]].plot(figsize = (15, 10), fontsize = 15)
plt.title("Simple Contrarian Strategy", fontsize = 20)
plt.legend(fontsize = 15)
plt.show()
```

```python
annualized_risk(df3[["DJI_daily_return", "strategy_percent_return"]])
```

## More complex Strategies & Backtesting vs. Fitting

-__More Complex Contrarian__ (Strategy to be tested): <br>
__a) Short Selling (-1)__ DJI tomorrow if today´s return was __greater than 0.01__ <br>
__b) Investing (+1)__ into DJI tomorrow if today´s return was __less than or equal to -0.1__

```python
df4 = df2.copy()
df4
```

```python
# switch position to short if that day's increase is at least 0.01

df4["position"] = np.where(df4["DJI_daily_return"] > 0.01, -1, 1)
df4
```

```python
df4["strategy_percent_return"] = df4["position"].shift() * df4["DJI_daily_return"]
df4
```

```python
df4["strategy_value_return"] = df4.strategy_percent_return.add(1, fill_value = 0).cumprod() * df4.iloc[0, 0]
df4
```

```python
df4[["DJI_close", "strategy_value_return"]].plot(figsize = (15, 10), fontsize = 15)
plt.title("Simple Complex Contrarian Strategy", fontsize = 20)
plt.legend(fontsize = 15)
plt.show()
```

```python
annualized_risk(df4[["DJI_daily_return", "strategy_percent_return"]])
```

Even if a Strategy __seems to outperform__ the basic contrarian strategy, the following issues need to be considered/analyzed as well: <br>
- Backtesting vs. Fitting -> Was the strategy __fitted__ and optimized on __historical data__? __Forward Testing__ is required!<br>
- __Transaction Costs__ -> Changing Positions trigger Cost. Include Costs!<br>
- __Tax effects__ -> Changing Positions can trigger (earlier) Tax Payments. Include Taxes!
